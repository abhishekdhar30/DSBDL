```python
df=df.replace(['<=50K','>50K'],[0,1])
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler1 = StandardScaler()
```

```python
from sklearn.preprocessing import LabelEncoder
labelEnc = LabelEncoder()
```

```python
df['sex'] = labelEnc.fit_transform(df['sex'])
df['sex']
```

```python
from sklearn.naive_bayes import GaussianNB
m1 = GaussianNB()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['age','sex','education.num']]
```

```python
y = df['income']
```

```python
x
```

```python
y
```

```python
x_tr, x_te, y_tr, y_te = train_test_split(x,y,test_size=0.25)
```

```python
x_tr.mean()
```

```python
x_tr.std()
```

```python
x_tr = scaler1.fit_transform(x_tr)
x_tr.std()
```

```python
x_te.mean()
```

```python
x_te = scaler1.transform(x_te)
x_te.mean()
```

```python
m1.fit(x_tr, y_tr)
```

```python
ypred = m1.predict(x_te)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(ypred, y_te)
```

```python
from sklearn.linear_model import LinearRegression
m2 = LinearRegression()
```

```python
m2.fit(x_tr, y_tr)
```

```python
ypred = m2.predict(x_te)
```
